```python
import numpy as np
import matplotlib.pyplot as plt
import math

import random
from statistics import mode


from scipy.stats import norm
```

```python
alpha = 1.5

xmin = 1

data = []
for i in range(1000):
    u = random.random()
    x = xmin / (u ** (1/alpha))
    data.append(x)
plt.figure(figsize=(5,3), dpi=120)

plt.hist(data)

plt.show()
```

```python
randomDraws = []
for i in data:
	randomDraws.append(random.choice(data))

plt.figure(figsize=(10,3), dpi=120)
plt.plot(range(len(randomDraws)), randomDraws)
plt.xlabel('time')
plt.ylabel("observation")
plt.title("Random draws from a Pareto distributed population (α=1.5)")

plt.axis('off')
plt.gca().set_position([0, 0, 1, 1])
plt.savefig("typeTwo_randDraws.svg")

plt.show()

print(mode(data))

```
