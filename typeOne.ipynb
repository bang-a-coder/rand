```python
import numpy as np
import matplotlib.pyplot as plt
import math

import random

from scipy.stats import norm
```

```python
mean = 100
std_dev = 3  

data = norm.rvs(loc=mean, scale=std_dev, size=1000)
 
plt.hist(data)

plt.show()
```

```python
randomDraws = []
for i in data:
	randomDraws.append(random.choice(data))

plt.figure(figsize=(10,3), dpi=120)
plt.plot(range(len(randomDraws)), randomDraws)
plt.xlabe = "time"
plt.ylabel = "observation"

plt.show()
```
